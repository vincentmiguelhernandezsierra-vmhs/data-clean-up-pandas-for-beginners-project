- Calcula el promedio de las calificaciones.

```python
calificaciones = [8.5, 9.2, 7.8, 8.9, 10]
promedio = sum(calificaciones) / len(calificaciones)
print("el promedio es:", promedio)
```

#### Ejercicio 01. ¿Cuál es la casa más cara en todo el dataset? (★☆☆)

Imprime la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle del Prado, Nº20 es la más cara y su precio es de 5000000 USD`

```python
precio_max = ds["price"].max()
print(precio_max)

casa_mas_cara = ds [ds["price"] == precio_max]
print(casa_mas_cara[["url_inmueble", "price"]])
```

#### Ejercicio 07. ¿Cuál es la media de precios en la población de "Arroyomolinos (Madrid)"? (★★☆)

Imprima el valor obtenido de la columna level5.

```python
filtro_arroyomolinos = ds["level5"] == "Arroyomolinos (Madrid)"
precios_arroyomolinos = ds.loc[filtro_arroyomolinos, "price"]

media_precios_arroyomolinos = precios_arroyomolinos.mean()

print("La media de precios en Arroyomolinos (Madrid) es", media_precios_arroyomolinos)
```

#### Ejercicio 08. Trazar el histograma de los precios para la población de "Arroyomolinos (Madrid)" y explica qué observas (★★☆)

Imprime el histograma de los precios y escribe en la celda del Markdown un breve análisis del trazado.


```python
import matplotlib.pyplot as plt

# Trazar el histograma de precios

precios_arroyomolinos = ds.loc[filtro_arroyomolinos, "price"].dropna()
precios_arroyomolinos = ds.loc[filtro_arroyomolinos, "price"].dropna()

plt.figure(figsize=(8, 5))
plt.hist(precios_arroyomolinos, bins=30)
plt.xlabel("precio de la vivienda")
plt.ylabel("Número de viviendas")
plt.title("Histograma de precios en Arroyomolinos (Madrid)")
plt.show()
```
